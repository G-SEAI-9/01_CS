```python
# Eine Klasse ist ein Bauplan für Objekte.
# Hier definieren wir den Bauplan für einen Mixer (Blender).
class Blender:

    # Klassen verkapseln Daten und Funktionalität
    # Klassenvariable: Alle Blender-Objekte haben diesen Container
    things_in_container = ["apple", "banana"]

    # Eine Methode ist eine Funktion, die zu einer Klasse gehört.
    # 'self' ist der Verweis auf das konkrete Objekt, das die Methode aufruft.
    def blend(self):
        result = ""

        # self.things_in_container greift auf die Liste dieses Objekts zu
        for word in self.things_in_container:
            result += word

        return result
```

```python
# Mit Klasse() erstellen wir eine Instanz — ein konkretes Objekt nach dem Bauplan.
my_blender = Blender()

# Wir können beliebig viele unabhängige Instanzen erstellen.
second_blender = Blender()

# Mit Punkt-Notation greift man auf Attribute und Methoden eines Objekts zu.
print(my_blender.things_in_container)
mix = my_blender.blend()

print(mix)
```

```python
class Blender:

    # __init__ ist der Konstruktor: er wird automatisch aufgerufen,
    # wenn ein neues Objekt erstellt wird (z.B. Blender([...])).
    # 'self' ist das neu erstellte Objekt, 'fruits' ein Parameter.
    def __init__(self, fruits):
        # Instanzvariable: gehört nur zu diesem einen Objekt, nicht zur Klasse.
        # Jede Instanz hat ihre eigene 'things_in_container'-Liste.
        self.things_in_container = fruits

    def blend(self):
        return "".join(self.things_in_container)

    # Methode, um Frucht in Mixer zu packen
    def append(self, new_fruit):
        # Validation: isinstance() prüft, ob new_fruit wirklich ein String ist.
        # So schützen wir die Liste vor ungültigen Werten (z.B. Zahlen).
        if isinstance(new_fruit, str):
            self.things_in_container.append(new_fruit)

    # __str__ ist eine sogenannte Dunder-Methode (double underscore).
    # Python ruft sie automatisch auf, wenn man print(objekt) schreibt.
    # Andere Dunder-Methoden bestimmen das Verhalten der Objekte in bestimmten Umständen, z.B. wie der + Operator mit dem Objekt funktioniert.
    def __str__(self):
        return f"This Blender contains: {", ".join(self.things_in_container)}"


my_blender = Blender(["kiwi", "orange"])
print(my_blender.things_in_container)

my_blender.append("mango")
my_blender.append("strawberry")
# 42 ist kein String → isinstance() schlägt an, nichts wird hinzugefügt
my_blender.append(42)
my_blender.blend()
```

```python
class Present:

    def __init__(self, name, price, url):
        # Eingabevalidierung: Preise müssen floats sein (z.B. 9.99, nicht "hello")
        if not isinstance(price, float):
            raise ValueError("Price must be a float")
        # Alle übergebenen Werte werden als Instanzvariablen gespeichert
        self.name = name
        self.price = price
        self.url = url
        # Instanzvariablen können auch mit einem Default initialisiert werden:
        self.person = ""

    def wrap(self):
        print(f"Wrapping this {self.name} in gift paper")

    def set_url(self, new_url):
        # Setter-Methode: kontrollierter Zugriff auf self.url.
        # Statt marble_track.url = 42 direkt zu erlauben, prüfen wir hier den Typ.
        if not isinstance(new_url, str):
            raise ValueError("URL must be a string")
        self.url = new_url

    def add_person(self, person_name):
        # Methode, um nachträglichen dem Geschenk eine Person zuzuweisen
        self.person = person_name
```

```python
# Klassenkonstruktoren können mit Positionalen und Keyword-Argumenten aufgerufen werden.
marble_track = Present(name="Kugelbahn", price=42.95, url="marb.le")

marble_track.wrap()

# person wurde im __init__ mit "" initialisiert — leer, bis add_person aufgerufen wird
print(marble_track.person)
marble_track.add_person("Miriam")
print(marble_track.person)


print(marble_track.url)
# Direktes Setzen wäre möglich, würde aber Validierung umgehen:
# marble_track.url = 42

# set_url() stellt sicher, dass nur Strings als URL gesetzt werden können
marble_track.set_url("new-marb.le")
print(marble_track.url)
```

```python
# Vererbung: BirthdayPresent erbt alle Attribute und Methoden von Present.
# Present heißt hier die Elternklasse (auch: Basisklasse).
class BirthdayPresent(Present):

    def __init__(self, name, price, url, date):
        # super().__init__() ruft den Konstruktor der Basisklasse auf.
        # So müssen wir name, price, url nicht nochmal implementieren.
        super().__init__(name, price, url)
        # Nur das neue Attribut 'date' kommt hinzu
        self.date = date

    # Neue Methode, die nur BirthdayPresent kennt
    def give(self):
        print(f"Giving {self.name} at {self.date}")

    # Method Overriding: wrap() aus der Elternklasse wird überschrieben.
    # Für ein Geburtstagsgeschenk soll eine andere Nachricht erscheinen.
    def wrap(self):
        print("This is a secret birthday present")
```

```python
bike = BirthdayPresent("Fahrrad", 321.31, "bike.com", "2026-07-01")

# BirthdayPresent hat kein __str__ → Python zeigt die Speicheradresse.
# Das wäre zu beheben, indem man __str__ erbt oder überschreibt.
print(bike)

# wrap() aus BirthdayPresent wird aufgerufen (überschriebene Version)
bike.wrap()

# give() ist eine neue Methode, die nur BirthdayPresent besitzt
bike.give()
```
