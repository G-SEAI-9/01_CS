```python
class Blender:

    def __init__(self, fruits):
        self.things_in_container = fruits

    def blend(self):
        # print(self)
        return "".join(self.things_in_container)

    # Methode, um Frucht in Mixer zu packen
    def append(self, new_fruit):
        if isinstance(new_fruit, str):
            self.things_in_container.append(new_fruit)

    def __str__(self):
        return f"This Blender contains: {", ".join(self.things_in_container)}"


my_blender = Blender(["kiwi", "orange"])
print(my_blender.things_in_container)

my_blender.append("mango")
my_blender.append("strawberry")
my_blender.append(42)
my_blender.blend()
```

```python
class Present:

    def __init__(self, name, price, url) -> None:
        if not isinstance(price, float):
            raise ValueError("Price must be a float")
        self.name = name
        self.price = price
        self.url = url
        self.person = ""

    def wrap(self):
        print(f"Wrapping this {self.name} in gift paper")

    def set_url(self, new_url):
        if not isinstance(new_url, str):
            raise ValueError("URL must be a string")
        self.url = new_url

    def add_person(self, person_name):
        self.person = person_name
```

```python
marble_track = Present(name="Kugelbahn", price=42.95, url="marb.le")

marble_track.wrap()

print(marble_track.person)
marble_track.add_person("Miriam")
print(marble_track.person)


print(marble_track.url)
# marble_track.url = 42

marble_track.set_url("new-marb.le")
print(marble_track.url)
```

```python
class BirthdayPresent(Present):

    def __init__(self, name, price, url, date):
        super().__init__(name, price, url)
        self.date = date

    def give(self):
        print(f"Giving {self.name} at {self.date}")

    def wrap(self):
        print("This is a secret birthday present")
```

```python
bike = BirthdayPresent("Fahrrad", 321.31, "bike.com", "2026-07-01")

print(bike)
bike.wrap()

bike.give()
```
