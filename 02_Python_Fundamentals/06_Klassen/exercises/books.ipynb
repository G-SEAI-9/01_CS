```python
# Create a Book Class
class Book:

    # __init__ wird automatisch aufgerufen, wenn ein Objekt erstellt wird.
    # 'self' verweist auf die gerade erstellte Instanz — so speichert jedes Objekt seine eigenen Werte.
    # 'read=False' ist ein Default-Parameter: wird er nicht angegeben, ist der Wert automatisch False.
    def __init__(self, name, author, release_date, read=False):
        self.name = name
        self.author = author
        self.release_date = release_date
        self.read = read

    # __str__ bestimmt, was ausgegeben wird, wenn man print() auf ein Objekt anwendet.
    def __str__(self):
        return f"{self.name} - written by {self.author}"
```

```python
# Objekte erstellen: Die Klasse wird wie eine Funktion aufgerufen — Python ruft dabei __init__ auf.
# Keyword-Argumente (name=...) machen den Code lesbarer, sind aber optional.
mybook1 = Book(
    name="Chroniken der Unterwelt", author="Cassandra Claire", release_date="2007"
)

# Hier werden Argumente positional übergeben — die Reihenfolge muss mit __init__ übereinstimmen.
# 'read=True' überschreibt den Default-Wert False.
lotr = Book("Lord of the Rings", "J.R.R. Tolkien", "1954", read=True)

print(mybook1.name)  # Zugriff auf ein Attribut über die Punkt-Notation
print(lotr.read)
print(mybook1)  # Ruft __str__ auf und gibt den formatierten String aus
```

```python
# Create a BookCollection Class


class BookCollection:

    # 'list_of_books=None' als Default statt [] — mutable Default-Argumente sind in Python eine häufige Fehlerquelle (-> Listen, Dictionaries)!
    def __init__(self, list_of_books=None):

        # Wenn eine Liste übergeben wurde, wird jeder Eintrag vorab geprüft.
        if list_of_books is not None:
            for entry in list_of_books:
                self._verify_book(entry)

        # 'or []' stellt sicher, dass self.list_of_books immer eine Liste ist, nie None.
        self.list_of_books = list_of_books or []

    # Der Unterstrich am Anfang (_verify_book) ist eine Konvention: "interne Methode, nicht von außen aufrufen".
    def _verify_book(self, thing):
        # isinstance() prüft, ob ein Objekt eine Instanz einer bestimmten Klasse ist.
        if not isinstance(thing, Book):
            raise ValueError(f"{thing} must be a book")

    def add_book(self, book):
        self._verify_book(book)  # Validierung vor dem Hinzufügen
        self.list_of_books.append(book)

    def mark_as_read(self, book_name):
        # Wir durchsuchen die Liste und ändern das Attribut direkt am Objekt.
        for book in self.list_of_books:
            if book.name == book_name:
                book.read = True
                return
        print(f"{book_name} not found")  # Feedback, wenn Buch nicht gefunden wurde

    def list_books(self):
        print(self)

    # __str__ für die Sammlung: baut einen formatierten Gesamtstring aus allen Büchern zusammen.
    def __str__(self):
        out = []
        for book in self.list_of_books:
            book_string = f"{book.name} - Read: {book.read}"
            out.append(book_string)
        return "\n".join(out)
```

```python
volles_regal = BookCollection([lotr, mybook1])

# volles_regal._verify_book("test")

print(volles_regal.list_of_books)
```

```python
mein_buecherregal = BookCollection()

print(mein_buecherregal.list_of_books)
```

```python
mein_buecherregal.add_book(lotr)
mein_buecherregal.add_book(
    Book(name="Qwert", author="Walter Moers", release_date="2025")
)
```

```python
mein_buecherregal.mark_as_read("Ronja Räubertochter")
mein_buecherregal.mark_as_read("Lord of the Rings")
mein_buecherregal.mark_as_read("Qwert")

for book in mein_buecherregal.list_of_books:
    print(f"{book.name} - read: {book.read}")
```
