```python
# Vergleichsoperatoren wie < > == geben immer einen Bool-Wert zurück: True oder False
5 < 10
```

```python
# Das Ergebnis eines Vergleichs wird in einer Variable gespeichert
condition = 15 < 10

# if prüft eine Bedingung. Ist sie True, wird der eingerückte Block ausgeführt.
# Die Einrückung (4 Leerzeichen) zeigt Python, welche Zeilen zum if-Block gehören. - Wichtig in einem Code-Block muss es immer die gleiche Einrückung geben.
if condition is True:
    print("This is True")
    print("This belongs to the same block")
    print("bla")

# else wird ausgeführt, wenn die if-Bedingung False ist — ein Auffangfall.
else:
    print("This is False")
```

```python
age = int(input("Gib dein Alter an: "))
has_drivers_license = True

# Mit 'and' müssen BEIDE Bedingungen True sein, damit der Block ausgeführt wird.
if age >= 18 and has_drivers_license:
    print("Darf Auto fahren")
# elif (else if) wird nur geprüft, wenn alle vorherigen Bedingungen False waren.
elif age >= 17:
    print("Darf begleitendes Fahren")
# else greift, wenn keine einzige Bedingung davor True war.
else:
    print("Darf kein Auto fahren")
```

```python
hour = 15

# Python prüft die Bedingungen von oben nach unten.
# Sobald eine Bedingung True ist, wird ihr Block ausgeführt und alle anderen übersprungen.
if hour < 12:
    print("Guten Morgen!")
elif hour < 14:
    print("Mahlzeit")
elif hour < 19:
    print("Guten Tag")
elif hour <= 24:
    print("N'Abend")
# else hier als Fehlerbehandlung: Dieser Fall sollte eigentlich nie eintreten.
else:
    print("ERROR: Stunde gibt's nicht")
```

```python
hour = 16
# Eine Variable wird mit einem Standardwert vorbelegt.
greeting = "Guten Morgen"

# Nur wenn die Bedingung zutrifft, wird der Wert überschrieben.
# Trifft sie nicht zu, bleibt der Standardwert erhalten.
if hour > 12:
    greeting = "Guten Tag"

print(greeting)
```

```python
hour = 15
# Ternary Operator: eine kompakte if/else-Zuweisung in einer Zeile.
# Aufbau: wert_wenn_true if bedingung else wert_wenn_false
greeting = "Guten Morgen" if hour < 12 else "Guten Tag"
print(greeting)
```

```python
# Bei vielen möglichen Werten wird if/elif schnell lang und unübersichtlich.
# Das ist ein guter Anlass, match/case zu verwenden (siehe nächste Zelle).
user_keyboard_input = "w"

if user_keyboard_input == "w":
    print("Move up")
elif user_keyboard_input == "a":
    print("Move left")
```

```python
user_keyboard_input = "w"

# match vergleicht einen Wert mit mehreren Mustern (cases).
# Dasselbe wie eine lange if/elif-Kette, aber übersichtlicher.
match user_keyboard_input:
    case "w":
        print("Move up")
    case "a":
        print("Move left")
    case "s":
        print("Move down")
    case "d":
        print("Move right")
    # Der Unterstrich _ ist der Wildcard-Fall: er passt auf alles Übrige.
    # Er entspricht dem else in einer if/elif-Kette.
    case _:
        print("DEFAULT")
```

```python
day = "Dienstag"

# Mit 'or' reicht es, wenn EINE der Bedingungen True ist.
if day == "Samstag" or day == "Sonntag":
    print("Wochenende")
else:
    print("Werktags")
```

```python
day = "Sonntag"

match day:
    # Der | Operator in match/case bedeutet "oder" — ein case kann mehrere Werte abdecken.
    # Das ist kürzer als im if-Konstrukt.
    case "Samstag" | "Sonntag":
        print("Wochenende")
    case _:
        print("Werktag")
```
