```python
hour = 3

if hour < 12:
    print("Guten Morgen!")
elif hour < 14:
    print("Mahlzeit")
elif hour < 19:
    print("Guten Tag")
elif hour <= 24:
    print("N'Abend")
else:
    print("ERROR: Stunde gibt's nicht")
```

```python
hour = 16
greeting = "Guten Morgen"

if hour > 12:
    greeting = "Guten Tag"

print(greeting)
```

```python
hour = 15
greeting = "Guten Morgen" if hour < 12 else "Guten Tag"  # Ternary Operator
print(greeting)
```

```python
user_keyboard_input = "w"

if user_keyboard_input == "w":
    print("Move up")
elif user_keyboard_input == "a":
    print("Move left")
```

```python
user_keyboard_input = "w"

match user_keyboard_input:
    case "w":
        print("Move up")
    case "a":
        print("Move left")
    case "s":
        print("Move down")
    case "d":
        print("Move right")
    case _:
        print("DEFAULT")
```

```python
day = "Dienstag"

if day == "Samstag" or day == "Sonntag":
    print("Wochenende")
else:
    print("Werktags")
```

```python
day = "Sonntag"

match day:
    case "Samstag" | "Sonntag":
        print("Wochenende")
    case _:
        print("Werktag")
```
