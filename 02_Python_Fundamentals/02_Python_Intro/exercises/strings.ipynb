```python
# Step 2: Access Characters and Slice Strings
print(first_name[0])
print(last_name[-1])
print(bio[:10])


some_str = "0123456789"

print(some_str[1:9:2])
print(some_str[9:1:-2])
```

```python
# Step 5: Check Substrings
print("pYtHo".lower() in bio.lower())
print("Java" not in bio)
```
