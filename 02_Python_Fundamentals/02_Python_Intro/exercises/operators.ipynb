```python
# Step 1: Arithmetic Operators


# Perform arithmetic operations

a = 15
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a**b)
```

```python
# Step 2: Assignment Operators
x = 10

# Modify x using assignment operators

x += 3
print(x)
x -= 3
x *= 3
x /= 3
x //= 3
```

```python
print(q := 1)
```

```python
# Step 3: Comparison Operators
# Compare a and b

print(a < b)
print(a > b)
print(a >= b)
print(a <= b)
print(10 >= 10)
print(10 > 10)
```

```python
print(a == b)
print(a != b)
```

```python
# Step 4: Logical Operators
is_python_fun = True
is_java_fun = False
# Combine Boolean variables

print(is_python_fun and is_java_fun)
print(is_python_fun or is_java_fun)

print(not (is_python_fun and is_java_fun))

print(not is_java_fun)
```

```python
# Step 5: Identity Operators
list1 = [1, 2, 3]
list2 = list1
list3 = [1, 2, 3]
# Check identities
print(list1 is list2)

# Check if list1 is list3
print(list1 is list3)  # Unterschiedliche Orte im RAM

print(list1 == list2)
print(list1 == list3)  # Gleiche Inhalte
```

```python
# Step 6: Membership Operators
text = "Python programming is fun!"
# Check membership
print("Python" in text)
```
