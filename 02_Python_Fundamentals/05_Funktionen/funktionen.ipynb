```python
# Diesen Block müssen wir jedes Mal neu schreiben, wenn wir ihn brauchen.
# → Wiederholung im Code ist fehleranfällig und schwer zu warten.
print(">-----------<")
print("Hello, World!")
print("<----------->")
```

```python
# Schon wieder dieselben drei Zeilen — genau das Problem, das Funktionen lösen.
print(">-----------<")
print("Hello, World!")
print("<----------->")
```

```python
# Eine Funktion wird mit 'def' definiert, gefolgt vom Namen und Klammern.
# Der eingerückte Block darunter ist der Funktionskörper — er wird erst ausgeführt,
# wenn die Funktion aufgerufen wird, nicht schon bei der Definition.
def print_with_format():
    print(">-----|-----<")
    print("Hello, World!")
    print("<-----|----->")


# Diese Zeile wird sofort ausgeführt — die Funktion selbst noch nicht.
print("Nach der Funktionsdefinition!")
```

```python
# Funktionsaufruf: Name + Klammern. Python springt in den Funktionskörper,
# führt ihn aus und kehrt dann hierher zurück.
print_with_format()
# print_with_format()
# print_with_format()
# print_with_format()
```

```python
# Funktionen lassen sich überall aufrufen — Python springt nach dem Aufruf wieder zurück in die Zeile.
for i in range(10):
    if i % 3 != 0:
        print(i)
    if i == 4:
        # Beim vierten Durchlauf wird die Funktion mitten in der Schleife aufgerufen.
        print_with_format()
    print("Hey")
```

```python
# Globale Variable — außerhalb der Funktion definiert, aber innerhalb sichtbar.
my_favorite_num = 42


# 'name' ist ein Parameter: ein Platzhalter für einen Wert, der beim Aufruf übergeben wird.
def print_greeting_with_format(name):
    # Globale Variablen können innerhalb einer Funktion gelesen werden.
    print(my_favorite_num)
    print(">-----|-----<")
    # f-String: der Wert von 'name' wird direkt in den Text eingesetzt.
    print(f"Hello, {name}!")
    print("<-----|----->")


# 'name' existiert nur innerhalb der Funktion — dieser Aufruf würde einen Fehler ergeben.
# print(name)
```

```python
# Argumente: die konkreten Werte, die beim Aufruf für den Parameter eingesetzt werden.
# Jeder Aufruf bekommt seinen eigenen 'name' — die Funktion ist wiederverwendbar.
# name = "Miguel"
print_greeting_with_format("Miguel")
print_greeting_with_format("Elise")
```

```python
# Funktionen können auch einen Wert berechnen und zurückgeben.
# 'number1' und 'number2' sind Parameter — sie erhalten beim Aufruf konkrete Werte.
def add(number1, number2):

    result = number1 + number2

    # 'return' gibt den Wert zurück an den Aufrufer und beendet die Funktion sofort.
    # Alles nach 'return' wird nie ausgeführt.
    return result
    print("Heyho")  # Diese Zeile ist unerreichbar.
```

```python
# Der Rückgabewert kann in einer Variable gespeichert und weiterverwendet werden.
sum_13_42 = add(13, 42)

print(sum_13_42)
```

```python
# Verschachtelte Aufrufe: der innere Aufruf wird zuerst ausgeführt (ergibt 10),
# dann wird add(5, 10) berechnet — Funktionen können direkt als Argumente dienen.
add(5, add(3, 7))
# add( 5,  10 )
```

```python
# Eine Funktion kann mehrere Werte als Tuple zurückgeben — einfach mit Komma trennen.
def mod_and_divide(num1, num2):
    mod = num1 % num2
    div = num1 // num2
    return (mod, div)
```

```python
# Tuple Unpacking: die zwei Rückgabewerte werden direkt in zwei Variablen entpackt.
days, years = mod_and_divide(1000, 365)

print(days)
print(years)

# Python hat genau so eine eingebaute Funktion:
div, mod = divmod(1000, 365)
```

```python
# Mehrere `return` sind möglich; beim ersten `return` wird die Funktion sofort verlassen.
def is_dividable_by_5(num):
    if num % 5 == 0:
        return "teilbar durch 5"
    else:
        return "nicht teilbar durch 5"


print(is_dividable_by_5(13))
print(is_dividable_by_5(15))
```

```python
# Funktionen können Fehler abfangen, bevor sie das Programm zum Absturz bringen.
# try/except erlaubt es, auf verschiedene Fehlertypen unterschiedlich zu reagieren.
def divide(num1, num2):

    try:
        result = num1 / num2
        return result

    # Falscher Datentyp (z.B. String statt Zahl)
    except TypeError:
        print("Both arguments need to be numbers")
        return None
    # Division durch Null ist mathematisch undefiniert
    except ZeroDivisionError:
        print("Never divide by Zero")
        return None
    # Auffangnetz für alle anderen unerwarteten Fehler
    except:
        print("An error occured")
        return None
```

```python
# Normaler Fall: gibt 25.0 zurück
print(divide(100, 4))

# Division durch 0 — löst ZeroDivisionError aus, Funktion gibt None zurück
print(divide(100, 0))

# Falscher Typ — String und Tuple sind keine Zahlen, löst TypeError aus
print(divide("GREMLINS", (True, False)))
```

```python
# Type Annotations: 'int | float' beschreibt, welchen Typ ein Parameter erwartet.
# '-> float | None' beschreibt den Rückgabetyp. Python erzwingt das nicht,
# aber es macht den Code lesbarer und ermöglicht bessere Editor-Unterstützung.
def divide(num1: int | float, num2: int | float) -> float | None:
    """
    Divides two numbers.
    """
    # Docstring (oben): kurze Beschreibung der Funktion — erscheint z.B. als Tooltip.
    try:
        result = num1 / num2
        return result

    except TypeError:
        print("Both arguments need to be numbers")
        return None
    except ZeroDivisionError:
        print("Never dividy by Zero")
        return None
    except:
        print("An error occured")
        return None
```

```python
# Testen der Version mit Type Annotations — verhält sich identisch, Editor kann Warnung geben
divide(5, 3.3)
divide("GRMLINS", 5)
```
