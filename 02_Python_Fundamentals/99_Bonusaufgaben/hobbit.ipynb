`````python
# Lasst uns noch ein bisschen mehr das Arbeiten mit Strings und Schleifen üben.
# Hier haben wir den berühmten Anfang eines berühmten Romans

opening = """In a hole in the ground there lived a hobbit. Not a nasty, dirty, wet hole, filled with the ends of worms and an oozy smell, nor yet a dry, bare, sandy hole with nothing in it to sit down on or to eat: it was a hobbit-hole, and that means comfort."""
print(opening)
`````

`````python
# Aufgabe 1:
# - Bestimmt, wie viele Sätze in diesem Anfang stecken
sentences = opening.split(". ")
clean_sentences = []
for sentence in sentences:
    if sentence.strip() != "":
        clean_sentences.append(sentence)

print("Anzahl der Sätze:")
print(len(clean_sentences))
print(sentences)
`````

`````python
# print(sum(opening.count(c) for c in ".!?"))
print(sum(opening.count(c) for c in [". ", "! ", "? "]))
`````

`````python
# Aufgabe 2:
# - Bestimmt, wie viele Wörter dieser Anfang insgesamt hat

# clean_text = opening.lower()
# clean_text = clean_text.replace(".", "")
# clean_text = clean_text.replace(",", "")
# clean_text = clean_text.replace(":", "")

clean_text = opening.lower().replace(".", "").replace(",", "").replace(":", "").split()

# words = clean_text.split()

print(len(clean_text))
`````

`````python
clean_text = opening.lower()

for zeichen in ".,:":
    clean_text = clean_text.replace(zeichen, "")

wörter = clean_text.split()
print(len(wörter))
`````

`````python
clean_text = "".join([c.lower() for c in opening if c not in "!?.,:;"]).split()

print(clean_text)
`````

`````python
# - Jetzt wollen wir die Worthäufigkeit dieses Textes wissen:
#     - Erstellt ein Dictionary word_frequency
word_frequency = {}

#     - Wandelt opening in eine Liste mit einzelnen Wörtern um. Loopt über die Liste und zählt, wie oft jedes Wort vorkommt
#     - Stellt sicher, dass alle Wörter als Kleinbuchstaben (lower case) behandelt werden, und entfernt alle Satzzeichen (".", ",", ":")
#     - Sammelt die Häufigkeit jedes Wortes in word_frequency, z.B.:
# word_frequency = {
#   "in": 3,
#   "a": ...
# }

for word in clean_text:  # O(n)
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1

print(word_frequency)
`````

`````python
word_frequency = {}
for word in clean_text:
    word_frequency[word] = clean_text.count(word)  # O(n²)

print(word_frequency)
`````

`````python
word_frequency = {}
for word in clean_text:
    curr_count = word_frequency.setdefault(word, 0)
    word_frequency[word] += 1

print(word_frequency)
`````

`````python
# Aufgabe 3
#  - Gebt euer Dictionary aus in einem lesbaren Format aus.
#  - Welches ist das häufigste Wort im Text?

highest_count = 0
most_common_word = ""

for word, count in word_frequency.items():
    if count > highest_count:
        most_common_word = word
        highest_count = count

print(f"{most_common_word}: {highest_count}")
`````
