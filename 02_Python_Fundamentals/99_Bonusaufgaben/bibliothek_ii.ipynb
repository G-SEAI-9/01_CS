Bonus Bonus

Erstellt für jeden Nutzer solch eine Email. Nutzt eine Schleife oder eine List Comprehension.


```python
def create_email(user):
    # Sprache des Benutzers holen, z.B. "en", "de" oder "es".
    lang = user["lang"]

    # Die Textteile für diese Sprache aus dem verschachtelten Dictionary holen.
    greeting = email_pieces["greeting"][lang]
    info = email_pieces["info"][lang]
    confirm = email_pieces["email_confirmation"][lang]
    borrowed = email_pieces["borrow_status_one"][lang]
    book_text = email_pieces["borrow_status_two"][lang]
    # thanks = email_pieces["thanks"][lang]
    outro = email_pieces["outro"][lang]

    # Datum für die jeweilige Sprache formatieren.
    formatted_date = user["return_date"]

    # Für Spanisch steht im Beispiel nach der Begrüßung ein Doppelpunkt.
    # Für Englisch und Deutsch steht ein Komma.
    if lang == "es":
        greeting_line = f"{greeting} {user['name']}:"
        # ending = f"{thanks}.\n{outro}."
    else:
        greeting_line = f"{greeting} {user['name']},"
        # ending = f"{thanks},\n{outro}"

    # Der fertige E-Mail-Text wird als mehrzeiliger f-String zusammengesetzt.
    email = f"""{greeting_line}
{info}.
{confirm}: {user['email']}.
{borrowed} {user['borrowed_books']} {book_text} {formatted_date}.
{outro}"""

    # Die fertige Mail zurückgeben, damit wir sie später printen oder speichern können.
    return email


for user in library_users:
    print(create_email(user))
```
