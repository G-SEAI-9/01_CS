Schreibt eine Funktion, um Reihen, Spalten und Diagonalen gemeinsam zu überprüfen.


```python
# Funktion zum Statuscheck:


def check_win(board, sign):
    [a, b, c], [d, e, f], [g, h, i] = board

    # Reihen
    if a == b == c == sign:
        return True
    if d == e == f == sign:
        return True
    if g == h == i == sign:
        return True
    # Spalten
    if a == d == g == sign:
        return True
    if b == e == h == sign:
        return True
    if c == f == i == sign:
        return True
    #  Diagonalen
    if a == e == i == sign:
        return True
    if c == e == g == sign:
        return True

    return False
```

```python
print(check_win(tic_tac_toe_board, "x"))
print(check_win(tic_tac_toe_board, "o"))
```

```python
def check_winner(board, player):
    for row in board:
        if row[0] == row[1] == row[2] == player:
            return True

    for column in range(3):
        if (
            (board[0][column]) == player
            and (board[1][column]) == player
            and (board[2][column]) == player
        ):
            return True

    if board[0][0] == player and board[1][1] == player and board[2][2] == player:
        return True

    if board[0][2] == player and board[1][1] == player and board[2][0] == player:
        return True

    return False


print(check_winner(tic_tac_toe_board, "x"))
print(check_winner(tic_tac_toe_board, "o"))
```

```python
def check_line(zelle):
    for player in ["x", "o"]:
        if all(i == player for i in zelle):
            print(f"Treffer für {player}")


def check_winner(tic_tac_toe_board):
    size = len(tic_tac_toe_board)

    # Zeilen
    for row in tic_tac_toe_board:
        check_line(row)

    # Spalten
    for col in range(size):
        check_line([row[col] for row in tic_tac_toe_board])

    # Hauptdiagonale
    check_line([tic_tac_toe_board[i][i] for i in range(size)])

    # Gegendiagonale
    check_line([tic_tac_toe_board[i][size - 1 - i] for i in range(size)])


check_winner(tic_tac_toe_board)
```

Schreibt eine Funktion, um einen ein `x` oder ein `o` in das Spielfeld einzutragen.
Was gibt es zu beachten - welche edge cases fallen euch auf?


```python
# Funktion für Spielerzug
def make_move(board, row, column, player):
    if row < 0 or row > 2 or column < 0 or column > 2:
        print("Ungültige Position")
        return False

    if board[row][column] != " ":
        print("Feld ist schon belegt")
        return False

    board[row][column] = player
    return True


for row in tic_tac_toe_board:
    print(row)

make_move(tic_tac_toe_board, 0, 1, "o")

for row in tic_tac_toe_board:
    print(row)
```

Schreibt eine Funktion, um das Spielfeld zurückzusetzen, indem alle Zellen auf " " gesetzt werden.


```python
def reset_board():
    # Gibt ein komplett leeres 3x3 Feld zurück
    return [[" ", " ", " "], [" ", " ", " "], [" ", " ", " "]]


tic_tac_toe_board = reset_board()
```

```python
def reset_board(board):
    for row in range(len(board)):
        for col in range(len(board[row])):
            board[row][col] = " "


reset_board(board=tic_tac_toe_board)

for row in tic_tac_toe_board:
    print(row)
```

bonus bonus:
Erstellt einen Spielloop:

> board -> Spielerzug `x` mit input -> Gewinnprüfung -> Spielerzug `o` mit input -> Gewinnprüfung -> ...


```python
from IPython.display import clear_output

reset_board(tic_tac_toe_board)
player = "x"

while True:
    # Spielfeld zeichnen
    clear_output()
    for row in tic_tac_toe_board:
        print(row)

    # User Input
    row = int(input(f"{player} - Row (0 - 2):"))
    col = int(input(f"{player} - Col (0 - 2):"))

    # User Spielzug validieren und durchführen
    success = make_move(board=tic_tac_toe_board, row=row, column=col, player=player)

    # wenn Spielerzug ungültig, wiederholen wir
    if not success:
        print("Invalid Move, try again")
        continue  # Spring an den Anfang der while-Schleife zurück

    # Gewinnbedingungen überprüfen
    if check_win(tic_tac_toe_board, player):
        clear_output()
        for row in tic_tac_toe_board:
            print(row)
        print(f"{player} won!")
        break  # while-Schleife beenden, wenn gewonnen

    # Gültiger Zug, aber nicht gewonnen -> nächster Spieler ist an der Reihe
    # if player == "x":
    #     player = "o"
    # else:
    #     player = "x"

    #  Alternative: Ternary
    player = "o" if player == "x" else "x"
```
