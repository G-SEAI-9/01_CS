```python
## Teil 1: Der erste Tag

# Löst die folgenden Aufgaben zunächst für einen einzelnen Tag:

# - Schritt 1: Konvertiert den langen String der CSV-Daten in eine list einzelner Zeilen. Ignoriert dabei den Header.

temp_days = berlin_may.splitlines()[1:]

print(temp_days)
```

```python
# - Schritt 2: Speichert den ersten Tag (2026-05-18) in einer eigenen Variable namens first_day, damit wir ihn genauer untersuchen können.
first_day = temp_days[0]
print(first_day)
```

```python
# - Schritt 3 (first_day): Konvertiert die Temperaturwerte dieses Tages in eine `list` von `floats`. Entfernt dabei das Datum am Ende.

first_day_list = first_day.split(",")[:-1]

temp_floats = [float(temp) for temp in first_day_list]

print(temp_floats)
```

```python
# - Schritt 4 (first_day): Findet einen Weg, die minimale und die maximale Temperatur zu ermitteln. Es gibt dafür praktische Built-in Functions in Python.

min_temp_first_day = min(temp_floats)
max_temp_first_day = max(temp_floats)

print(f"Niedrigste: {min_temp_first_day}")
print(f"Höchste: {max_temp_first_day}")
```

```python
# - Schritt 5 (first_day): Sagen wir, der Tag (Daylight Day) dauert immer von 6:00 bis einschließlich 18:00 Uhr. Berechnet die Durchschnittstemperatur für diesen Zeitraum.

first_day_daylight = temp_floats[6:19]

print(first_day_daylight)

average_temp_first_day = round(sum(first_day_daylight) / len(first_day_daylight), 2)

print(average_temp_first_day)
```

```python
# ## Teil 2: Der gesamte Zeitraum

# Jetzt wird es fortgeschrittener. Versucht so weit wie möglich ohne die Hilfe von KI zu kommen.

# - Schritt 6: Erstellt aus der Liste aller Zeilen eine neue verschachtelte Liste namens `temperatures_floats`. Diese soll für jeden Tag eine eigene Liste mit den reinen Temperaturwerten als `floats` enthalten (ohne die Datumsangaben am Ende).
#   Das Ergebnis sollte so aussehen:
#   [[15.4, 14.8, 14.2, ...], [16.1, 15.3, 14.9, ...], ...]
#   (Hinweis: Diese Aufgabe ist knifflig, da sie Nested Comprehensions erfordert. Beispiele zur Syntax findet ihr unter GeeksforGeeks. https://www.geeksforgeeks.org/python/nested-list-comprehensions-in-python/
#   Vielleicht findet ihr auch andere Lösungen für das Problem.
#   - Falls ihr hier stecken bleibt, poste ich im Laufe des Tages eine mögliche Lösung hier, damit ihr weiterarbeiten könnt.)
```

```python
temp_floats = [
    [float(val) for val in row.split(",")[:-1]] for row in berlin_may.splitlines()[1:]
]

print(temp_floats)
```

```python
# - Schritt 7: Erstellt aus der Liste temperatures_floats zwei neue Listen:
#   Eine Liste namens `maxima`, die nur die täglichen Maximum-Temperaturen enthält.
#   Eine Liste namens `minima`, die nur die täglichen Minimum-Temperaturen enthält.
minima = [min(temperatur) for temperatur in temp_floats]
maxima = [max(temperatur) for temperatur in temp_floats]


print(minima)
print(maxima)
```

```python
# - Schritt 8: Erstellt basierend auf `temp_floats` eine Liste, welche die Durchschnittstemperatur jedes Daylight Day (6:00–18:00 Uhr) enthält.

daylight_days = [day[6:19] for day in temp_floats]

daylight_avg = [round(sum(day) / len(day), 2) for day in daylight_days]

print(daylight_avg)
```

```python
# - Schritt 9: Findet die absolut höchste und niedrigste Temperatur des gesamten Zeitraums sowie das dazugehörige Datum heraus.
#   (Hierfür müsst ihr keine komplizierte List Comprehension schreiben! Ihr könnt das in mehreren Einzelschritten im Code lösen:
#   Tipps:
#   Erstellt zuerst eine separate Liste dates, die nur die Datums-Strings aller Tage enthält.
#   Ermittelt die höchste Temperatur aus eurer maxima-Liste.
#   Findet mit der Listen-Methode .index() heraus, an welcher Position (Index) dieser Höchstwert liegt.
#   Nutzt diesen Index, um das passende Datum aus eurer dates-Liste auszulesen.
#   Wiederholt das Ganze für die niedrigste Temperatur mit der minima-Liste.)

max_temp = max(maxima)
max_temp_idx = maxima.index(max_temp)

warmest_date = temp_days[max_temp_idx].split(",")[-1]
```

```python
# - Schritt 10: Findet heraus, welcher Tag (Datum) im Durchschnitt der wärmste Daylight Day war.
max_average = max(daylight_avg)
max_avg_idx = daylight_avg.index(max_average)

print(f"Wärmster Tag im Durchschnitt: {temp_days[max_avg_idx].split(",")[-1]}")
```
