```python
# Eine leere Liste mit eckigen Klammern erstellen
students = []
# Alternativ: list() ist die eingebaute Funktion — macht dasselbe
more_students = list()
print(more_students)
```

```python
# Eine Liste mit 5 Namen — alle Elemente in einer einzigen Variable
students = ["Michael", "Katharina", "Nicole", "Tomas", "Eric"]
print(students)
# type() zeigt den Datentyp — hier: list
type(students)
```

```python
# Achtung: 'list' ist ein reservierter Name in Python.
# Eine Variable so zu nennen überschreibt die eingebaute Funktion — das sollte man vermeiden!
# list = [1, 2, 3]
# print(list)
```

```python
# Auf einzelne Elemente zugreifen: Index in eckigen Klammern
# Python zählt ab 0 — das erste Element hat Index 0
print(students[0])  # erstes Element
print(students[1])  # zweites Element
print(students[2])
print(students[3])
# Negative Indizes zählen vom Ende der Liste
print(students[-1])  # letztes Element
print(students[-2])  # vorletztes Element
# students[99] würde einen IndexError auslösen, weil die Liste nur 5 Elemente hat
# print(students[99])
```

```python
students = ["Michael", "Katharina", "Nicole", "Tomas", "Eric"]

# Slicing: list[start:stop] — gibt Elemente von Index start bis stop-1 zurück
print(students[0:3])  # Index 0, 1, 2
print(students[:1])  # von Anfang bis Index 0
print(students[2:])  # ab Index 2 bis zum Ende

# Drittes Argument ist der Schritt: list[start:stop:step]
print(students[0:4:2])  # jeden zweiten Index von 0 bis 3
print(students[::2])  # jeden zweiten Index der ganzen Liste

# Negativer Schritt kehrt die Richtung um
print(students[3:1:-1])  # rückwärts von Index 3 bis 2

# [:] ohne Argumente kopiert die gesamte Liste — nützlich um eine unabhängige Kopie zu erstellen
copy_students = students[:]

print(copy_students)
```

```python
students = ["Michael", "Katharina", "Nicole", "Tomas", "Eric"]

# Element an einem bestimmten Index überschreiben
students[0] = "Micha"

print(students)

# del entfernt das Element an dem angegebenen Index dauerhaft aus der Liste
del students[0]

print(students)
```

```python
# .append() fügt ein Element ans Ende der Liste hinzu
students.append("Michael")
print(students)
new_one = "Gabriele"
# Die Liste wird direkt verändert (mutiert) — es wird keine neue Liste erstellt
students.append(new_one)  # mutating
print(students)
```

```python
# .pop() entfernt das letzte Element und gibt es zurück — man kann es in einer Variable speichern
last_element = students.pop()
print(last_element)  # das entfernte Element
print(students)  # die Liste ohne das letzte Element
```

```python
# .insert(index, wert) fügt ein Element an einer bestimmten Position ein
# Alle nachfolgenden Elemente rücken einen Platz nach hinten
students.insert(2, "Guybrush")

print(students)
```

```python
# .remove() sucht nach dem ersten Vorkommen des Wertes und entfernt ihn
# Im Gegensatz zu del braucht man nicht den Index zu kennen
students.remove("Guybrush")
print(students)
```

```python
# 'in' prüft, ob ein Wert in der Liste enthalten ist — gibt True oder False zurück
print("Eric" in students)
# 'not in' ist das Gegenteil
print("Jane" not in students)
```

```python
my_fav_var = "oranges"

# Listen können gemischte Datentypen enthalten: Zahlen, Booleans, Strings, Variablen — sogar andere Listen
my_desk = [1, True, "Coffee", my_fav_var, [1, 2, 3, 4]]

print(my_desk)
```

```python
# Eine 2D-Liste (Matrix): eine Liste, die selbst Listen enthält — gut für Tabellen oder Spielfelder
tic_tac_toe_board = [["o", "", ""], ["", "", "ö"], ["x", "", ""]]

# Zugriff: erst die Zeile, dann die Spalte
print(tic_tac_toe_board[1][2])  # board[row][col]
```

```python
# for-Schleife über eine Liste: name nimmt nacheinander den Wert jedes Elements an
for name in students:
    print(name)
```

```python
# Man kann auch direkt über eine Liste in der Schleife iterieren — ohne sie vorher in einer Variable zu speichern
for num in [1, 2, 3, 4]:
    print(num)
```

```python
numbers = [0, 1, 2, 3, 4, 5]

# List Comprehension: eine kompakte Schreibweise um eine neue Liste aus einer anderen zu erstellen
# Syntax: [ausdruck for element in liste]
doubled = [num * 2 for num in numbers]

# Das ist gleichbedeutend mit dieser ausführlichen Variante:
# for num in numbers:
#     doubled.append(num * 2)

print(doubled)
```
