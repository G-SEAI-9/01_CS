```python
my_dictionary = {
    "a": "Hello, World!",
    "b": 123,
    "c": (1, True),
    "d": [9, 8, 7, 6],
    0: "Tomas",
    1: "Michael",
    3.14: "PI",
    ("Guybrush", True): "User is active",
}
```

```python
print(my_dictionary["d"])
print(my_dictionary[0])
print(my_dictionary[3.14])
print(my_dictionary[("Guybrush", True)])
print(my_dictionary["z"])
```

```python
print(my_dictionary.get("z", "Default - z existiert nicht"))
```

```python
my_dictionary["y"] = "Etwas Neues"
print(my_dictionary["y"])
```

```python
print(my_dictionary.keys())
print(my_dictionary.values())
print(my_dictionary.items())
```

```python
copy_dict = my_dictionary.copy()
print(copy_dict)
```

```python
element_a = my_dictionary.pop("a")
print(element_a)
# print(my_dictionary["a"])
```

```python
last_inserted = my_dictionary.popitem()
print(last_inserted)
```

```python
del my_dictionary["b"]
print(my_dictionary)
my_dictionary.clear()
```
