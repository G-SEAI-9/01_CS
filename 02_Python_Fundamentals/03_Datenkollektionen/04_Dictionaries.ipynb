```python
# veränderbar     | geordnet   | index | Duplikate       | lineare Suche
my_lst = [1, 2, 3]
# unveränderbar   | geordnet   | index | Duplikate       | lineare Suche
my_tuple = (1, 2, 3)
# veränderbar     | ungeordnet | keys  | keine Duplikate | sofortige Abfrage
my_set = {1, 2, 3}

# Ein Dictionary kombiniert das Beste: veränderbar, schnelle Suche per Key, geordnet (seit Python 3.7 (!))
# veränderbar     | geordnet   | keys  | keine Duplikate Keys | sofortige Abfrage per Key
my_dict = {"key": "value"}
```

```python
# Dictionary erstellen: {key: value, key: value, ...}
# Keys müssen eindeutig sein, Values können beliebige Typen sein
english_to_german = {"morning": "Morgen", "noon": "Mittag", "evening": "Abend"}

# Zugriff über den Key in eckigen Klammern — wie bei einer Liste
print(english_to_german["morning"])

# Wert überschreiben: einfach neu zuweisen — der Key bleibt, der Value ändert sich
english_to_german["morning"] = 123
print(english_to_german["morning"])
```

```python
# Keys müssen hashbar (unveränderbar) sein: Strings, Zahlen und Tuples sind erlaubt.
# Values können alles sein: Strings, Zahlen, Listen, Tuples, andere Dicts, ...
my_dictionary = {
    "a": "Hello, World!",  # String als Key und Value
    "b": 123,  # Integer als Value
    "c": (1, True),  # Tuple als Value
    "d": [9, 8, 7, 6],  # Liste als Value
    0: "Tomas",  # Integer als Key
    1: "Michael",
    3.14: "PI",  # Float als Key
    (
        "Guybrush",
        True,
    ): "User is active",  # Tuple als Key — möglich, weil Tuples unveränderbar sind
}
```

```python
# Zugriff funktioniert mit jedem Key-Typ — Hauptsache der Key existiert
print(my_dictionary["d"])
print(my_dictionary[0])
print(my_dictionary[3.14])
print(my_dictionary[("Guybrush", True)])

# Nicht existierender Key → KeyError: Python wirft eine Fehlermeldung
print(my_dictionary["z"])
```

```python
# .get(key, default) — sicherer Zugriff ohne KeyError
# Wenn der Key nicht existiert, wird statt eines Fehlers der Default-Wert zurückgegeben, oder None
print(my_dictionary.get("z", "Default - z existiert nicht"))
```

```python
# Neuen Key-Value-Eintrag hinzufügen: einfach zuweisen — existiert der Key noch nicht, wird er angelegt
my_dictionary["y"] = "Etwas Neues"
print(my_dictionary["y"])
```

```python
# .keys()   → alle Keys als dict_keys (iterierbar, wie eine Liste)
print(my_dictionary.keys())

# .values() → alle Values als dict_values
print(my_dictionary.values())

# .items()  → alle Key-Value-Paare als Tuples — nützlich für Schleifen: for k, v in dict.items()
print(my_dictionary.items())
```

```python
# .copy() erstellt eine flache Kopie des Dictionarys.
# Änderungen an copy_dict beeinflussen my_dictionary nicht (und umgekehrt).
# (Achtung: gilt nicht für stärker verschachtelte Strukturen)
copy_dict = my_dictionary.copy()
print(copy_dict)
```

```python
# .pop(key) entfernt den Eintrag mit diesem Key und gibt seinen Value zurück.
# Nützlich, wenn man den Wert noch braucht, bevor man ihn löscht.
element_a = my_dictionary.pop("a")
print(element_a)
# print(my_dictionary["a"])  # würde jetzt KeyError werfen — der Key ist weg
```

```python
# .popitem() entfernt das zuletzt eingefügte Key-Value-Paar und gibt es als Tuple zurück.
last_inserted = my_dictionary.popitem()
print(last_inserted)
```

```python
# del entfernt einen bestimmten Key direkt — kein Rückgabewert (im Gegensatz zu .pop())
del my_dictionary["b"]
print(my_dictionary)

# .clear() löscht alle Einträge auf einmal — das Dictionary existiert danach noch, ist aber leer
my_dictionary.clear()
```

```python
# Praxisbeispiel: Liste von Dictionaries — typische Datenstruktur für z.B.  Datenbanken
# Jeder Tweet ist ein Dictionary
tweets = [
    {
        "author": "Wes Bos",
        "text": "This is my latest stuff...",
        "timestamp": 17878987176765,
    },
    {
        "author": "Guido van Rossum",
        "text": "Python is awesome",
        "timestamp": 17878987176723,
        "likes": ["Eric Schmitt", "Carl Sagan"],
    },
    {
        "author": "Jon Snow",
        "text": "This is my latest stuff...",
        "timestamp": 178789871234765,
    },
    {
        "author": "Wes Bos",
        "text": "This is my latest stuff...",
        "timestamp": 17878986546765,
    },
]

# Zugriff: erst Index für den Tweet in der Liste, dann Key im Dictionary
print(tweets[1]["text"])

# Die Likes-Liste lässt sich direkt weiterverarbeiten — hier alphabetisch sortieren
print(sorted(tweets[1]["likes"]))
```
