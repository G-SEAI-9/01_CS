````python
# - Erstellt auf Basis der vollständigen Liste eine neue mithilfe von List Comprehension. Die soll ausschließlich Benutzer enthalten, die Bücher ausgeliehen haben.

new_list = [user.strip().split(",") for user in lib_lst]

users_with_borrowed_books = [books for books in new_list if int(books[3]) > 0]

print(len(new_list))
print(len(users_with_borrowed_books))
````
