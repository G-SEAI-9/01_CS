```python
set_a = set(["Birne", "Banane", "Orange", "Brombeere"])
set_b = set(["Brombeere", "Erdbeere", "Ananas", "Mango", "Melone", "Zitrone", "Kiwi"])

print(set_a)
print(set_b)
print(set_a.isdisjoint(set_b))
print(set_a.issubset(set_b))
print(set_a.issuperset(set_b))
```

```python
print("Union", set_a.union(set_b))

# print( set_a | set_b )
```

```python
print("Intersection", set_a.intersection(set_b))

# print( set_a & set_b )
```

```python
print("Differenz", set_a.difference(set_b))
```

```python
print("Symmetrische Diff", set_a.symmetric_difference(set_b))
```

```python
set_a.difference_update(set_b)
print(set_a)
```

```python
sm_set = set([1, 2, 3])
lg_set = set([1, 2, 3, 4, 5, 6, 7])

print(sm_set.issubset(lg_set))
print(f"Ist large_set eine Obermenge von small_set? {lg_set.issuperset(sm_set)}")

print(sm_set.isdisjoint(lg_set))
print(set_a.isdisjoint(set_b))
```
