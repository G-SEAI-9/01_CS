Dictionary Übungen


```python
# 1. Create and Print a Dictionary

person_data = {"name": "Tomas", "age": 27, "city": "Hamburg"}

print(person_data)
```

```python
# 2. Access Dictionary Elements
print(person_data["name"])

print(person_data.get("email", "Nichts gefunden"))

print(person_data.keys())
print(person_data.values())
print(person_data.items())
```

```python
# 3. Check for Key Existence
if "age" in person_data:
    print("Key exists")
else:
    print("Key does not exist")

print("job" in person_data)
```

```python
print("Hamburg" in person_data.values())
```

```python
# 4. Change and Update Dictionary Elements
person_data["city"] = "Salvador"
print("Updated person_data with new city: ", person_data)
```

```python
# 5. Add New Items to the Dictionary
person_data.update({"Beruf": "Regisseur"})
print(person_data)
```

```python
# 6. Remove Items from the Dictionary
removed_value = person_data.pop("city", "Antlantis")
print("Entfernter Wert:", removed_value)
print(person_data)
```

```python
# 7. Copy a Dictionary

person_data["some_list"] = [1, 2, 3]

person_copy = person_data.copy()

person_copy["Beruf"] = "Software Engineer"
person_copy["some_list"][0] = "etwas anderes"

print(person_data)
print(person_copy)
```

```python
# 8. Using setdefault()
print(person_data.setdefault("age", 123))

print(person_data.setdefault("lieblings_eissorte", "schoko"))
```

```python
print(person_data)
person_data["lieblings_eissorte"] = "strawberry"
print(person_data)
```
