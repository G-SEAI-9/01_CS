```python
# Step 1: Create and Print a List
colors = ["türkis", "blau", "lila", "schwarz", "rosa"]
print(colors)
```

```python
# Step 2: Access Elements by Index and Negative Index
print("First Item:", colors[0])
print(colors[-1])
```

```python
# Step: 3 Slice a List
print(colors[1:4])
print(colors[2:])
print(colors[:3])
```

```python
# Step 4: Check if an Item Exists

if "grün" in colors:
    print("Grün kommt vor")
if "grün" not in colors:
    print("Grün kommt nicht vor")
```

```python
# Step 5: Add Items
colors.append("gelb")
print(colors)
```

```python
colors.insert(2, "grau")
print(colors)
```

```python
# Step 6: Change Items
colors[-1] = "sonnengelb"
print(colors)
colors[1:3] = ["himmelblau", "steingrau"]
print(colors)
```

```python
# Step 7: Remove items
colors.remove("steingrau")
print(colors)
last = colors.pop()
# colors.clear()
print(colors)
```

```python
# Step 8: Copy a list
copy_colors = colors.copy()
another_copy = colors[:]

# colors[0] = 42

print(copy_colors)
print(another_copy)
```

```python
# Step 9: Concatenate and Extend
print(colors + ["grün", "cyan"])
print(colors)
colors.extend(["pink", "mitternachtsblau", "pink"])
print(colors)
```

```python
# Step 10: Sort and Reverse
numbers = [4, 2, 5, 1, 3, 12, 10]

# numbers.reverse()
# numbers.sort()

sorted_numbers = sorted(numbers)

print(numbers)
print(sorted_numbers)


print(sorted(colors))
```

```python
# 11. Count and Index
print(colors.count("pink"))
print(colors.count("schwarz"))
print(colors.count("stegosaurus"))
anzahl_pink = colors.count("pink")
print(anzahl_pink)

print(colors)
position_mitternachtsblau = colors.index("mitternachtsblau")
print(position_mitternachtsblau)
```

```python
# List comprehension

capital_colors = [color.upper() for color in colors if color == "türkis"]

print(capital_colors)
```
