```python
# 1. Create a Tuple
my_tuple = ("a", 2, "b", 4, "c", 5)
```

```python
# 2. Print the Tuple
print(my_tuple)
```

```python
# 3. Access Tuple Items
print(my_tuple[0])
print(my_tuple[-1])
```

```python
# 4. Slice the Tuple
print(my_tuple[2:4])
print(my_tuple[:3])
print(my_tuple[3:])
```

```python
# 5. Check if an Item Exists
if 4 in my_tuple:
    print("4 is present in my_tuple")
else:
    print("4 is not in my_tuple")


print(2 in my_tuple)
```

```python
# 6. Count and Index
print(my_tuple.count("b"))
print(my_tuple.index("b"))
```

```python
# 7. Packing and Unpacking
note1, note2, note3, note4, note5, note6 = my_tuple

print(note3)
```

```python
note1, note2, *rest = my_tuple
print(note1)
print(note2)
print(rest)
```

```python
x, y, *_, last = my_tuple

print(last)
```

```python
# 8. Joining Tuples
another_tuple = ("x", "y", "z")
print(another_tuple + my_tuple)
print(another_tuple * 4)
```
