```python
my_lst = [1, 2, 3]

# Tupel werden mit runden Klammern erstellt — ähnlich wie Listen, aber unveränderlich
my_tuple = (1, 2, 3)

print(my_tuple)
print(type(my_tuple))

# Typischer Anwendungsfall: zusammengehörige Werte mit fester Bedeutung (hier: Koordinaten)
location = (53.4, 13.4)
```

```python
# Indexzugriff funktioniert genau wie bei Listen — Index beginnt bei 0
print(my_tuple[0])

print(location[1])
```

```python
# Tupel sind unveränderlich (immutable) — dieser Code löst einen TypeError aus
my_tuple[0] = "Etwas Neues"
```

```python
fruits1 = ("Apfel", "Banane", "Erdbeere")
fruits2 = ("Ananas", "Mango", "Papaya")

# + verkettet zwei Tupel zu einem neuen — die Originale bleiben dabei unverändert
combined = fruits1 + fruits2

print(combined)
print(fruits1)
# * wiederholt das Tupel n-mal und erzeugt ebenfalls ein neues Tupel
print(fruits1 * 3)
```

```python
# Ein-Element-Tupel braucht zwingend ein Komma nach dem Wert —
# ohne Komma würde Python (4) nur als geklammerten Ausdruck interpretieren, nicht als Tupel
single_val_tuple = (4,)

type(single_val_tuple)
```

```python
location = (53.4, 13.4)

# print(location[0])

# Unpacking: Tupelwerte werden in einem Schritt auf einzelne Variablen aufgeteilt.
# Die Anzahl der Variablen muss exakt mit der Länge des Tupels übereinstimmen.
latitude, longitude = location

print(latitude)
print(longitude)
```

```python
colors = ("mitternachtsblau", "steingrau", "sonnengelb", "persilweiß")

# print(colors)

# _ ist Konvention für "dieser Wert wird nicht gebraucht" — er wird ignoriert
color1, _, color3, _ = colors

# *rest (Stern-Ausdruck) sammelt alle übrig bleibenden Werte in einer Liste
_, interesting_color, *rest = colors

# print(len(colors))
# print(color3)
# print(_)

print(interesting_color)
print(rest)
```

```python
numbers_tup = (1, 2, 3, 4, 5, 6)

# *_ verwirft alle restlichen Werte — nützlich wenn man nur die ersten n Elemente braucht
eins, zwei, *_ = numbers_tup

print(eins)
print(zwei)
print(_)
```

```python
# Praxisbeispiel: Tupel eignen sich gut für Datensätze mit fester Struktur.
# Hier speichert jedes Tupel genau zwei zusammengehörige Werte: Name + Anwesenheit.
presence_thur = [("Michael", False), ("Tomas", True), ("Nicole", True)]

# List Comprehension mit Tuple Unpacking in der for-Zeile:
# (name, present) entpackt jedes Tupel direkt — kein Index-Zugriff nötig.
# Die if-Bedingung filtert auf Abwesende (present is False).
absentees = [name for (name, present) in presence_thur if present is False]

print(absentees)
```

```python
#  Konvertrierung von Liste -> Tupel -> Liste
my_lst = [1, 2, 3]

my_tup = tuple(my_lst)

list(my_tup)
```

```python
some_list = [1, 2, 3, 4, "TARGET", 5, 6, 7, 8, 3]

# .copy() ist die lesbare Variante
# copy_lst = some_list.copy()
# print(copy_lst)
# copy_lst[0] = "Hallo"
# print(copy_lst)   # Kopie verändert
# print(some_list)  # Original unverändert

# [:] (leerer Slice) erzeugt dasselbe Ergebnis wie .copy() — 'pythonic'
another_copy = some_list[:]
print(another_copy)
```

```python
# .index() sucht den Index des ersten Vorkommens — hier: wo steht "TARGET"?
index_target = some_list.index("TARGET")

# Slice von Anfang bis (ausschließlich) idx_tar gibt alle Elemente davor zurück
print(some_list[0:index_target])
```

```python
# .count() durchsucht die gesamte Liste und zählt alle Vorkommen von 3
# (some_list enthält 3 an zwei Stellen)
print(some_list.count(3))
```
