```python
my_list = [6, 5, 3, 7, 4, 6, 4, 3, 6, 7, 4, 6, 0]
my_tuple = (6, 5, 3, 7, 4, 6, 4, 3, 6, 7, 4, 6, 0)

# Ein Set wird mit geschweiften Klammern {} erstellt.
# Im Gegensatz zu Liste und Tuple speichert ein Set jeden Wert nur einmal —
# Duplikate werden automatisch entfernt.
# Außerdem hat ein Set keine feste Reihenfolge: die Ausgabe kann sich von der Eingabe unterscheiden.
my_set = {6, 5, 3, 7, 4, 6, 4, 3, 6, 7, 4, 6, 0}

print("Liste", my_list)
print("Tuple", my_tuple)
print("Set  ", my_set)  # Nur eindeutige Werte, in beliebiger Reihenfolge
```

```python
# Intern arbeitet ein Set mit sogenannten Hashes.
# hash() berechnet für einen Wert eine eindeutige Zahl — Python nutzt diese, um Elemente schnell zu finden.
# hash(1)

# Listen können nicht gehasht werden, weil sie veränderbar (mutable) sind.
# Ein Set braucht unveränderliche (immutable) Werte: Zahlen, Strings, Tuples (!) — aber keine Listen.
# hash([1, 2, 3])
```

```python
# Membership Check
# Bei einem Set ist diese Suche deutlich schneller als bei einer Liste,
# weil Python den Hash-Wert direkt nachschlagen kann — ohne alle Elemente durchzugehen.
print(0 in my_list)
print(0 in my_set)
```

```python
# set() wandelt eine Liste in ein Set um.
# "Civilization" kommt in der Liste zweimal vor — im Set erscheint es nur einmal.
good_old_games = set(
    [
        "Morrowind",
        "Civilization",
        "Dune",
        "Giana Sisters",
        "Oil Imperium",
        "Civilization",
    ]
)

print(good_old_games)
```

```python
# .add() fügt ein neues Element ins Set ein.
# Ist das Element bereits vorhanden, passiert nichts — keine Fehlermeldung, kein Duplikat.
good_old_games.add("Dune 2")

# .discard() entfernt ein Element, falls es im Set vorhanden ist.
# Ist es nicht enthalten, passiert ebenfalls nichts (kein Fehler).
# Alternative: .remove() wirft einen Fehler, wenn das Element fehlt.
good_old_games.discard("Giana Sisters")
print(good_old_games)
```

```python
set_1 = set([1, 2, 3, 4, 5])
set_2 = set([4, 5, 6, 7, 8])

# union: alle Elemente aus beiden Sets (Vereinigung) → {1,2,3,4,5,6,7,8}
print("union               ", set_1.union(set_2))

# intersection: nur Elemente, die in beiden Sets vorkommen (Schnittmenge) → {4,5}
print("intersection        ", set_1.intersection(set_2))

# difference: Elemente aus set_1, die nicht in set_2 sind (Differenz) → {1,2,3}
print("difference          ", set_1.difference(set_2))

# symmetric_difference: Elemente, die in genau einem der beiden Sets sind (symmetrische Differenz) → {1,2,3,6,7,8}
print("symmetric difference", set_1.symmetric_difference(set_2))

# Alle Mengenoperationen erzeugen ein neues Set — set_1 bleibt unverändert.
print(set_1)
```

```python
colors = set(["green", "orange", "navy", "purple"])
# other_colors = set(["black", "blue", "white"])
other_colors = set(["navy", "green"])

# issubset: prüft, ob other_colors eine Teilmenge von colors ist
# → True, weil "navy" und "green" beide in colors enthalten sind
print(other_colors.issubset(colors))

# issuperset: prüft das Gegenteil — ob colors alle Elemente von other_colors enthält
print(colors.issuperset(other_colors))

# isdisjoint: prüft, ob beide Sets keine gemeinsamen Elemente haben
# → False, weil "navy" und "green" in beiden vorkommen
print(other_colors.isdisjoint(colors))
```

```python
emails = [
    "a.mail.com",
    "b.mail.com",
    "d.mail.com",
    "f.mail.com",
    "a.mail.com",
    "b.mail.com",
    "g.mail.com",
]

# Typischer Anwendungsfall: Duplikate aus einer Liste entfernen.
# set() entfernt alle Dopplungen, list() wandelt das Ergebnis wieder in eine Liste um.
# -> Die ursprüngliche Reihenfolge ist danach nicht mehr garantiert.
unique_mails = list(set(emails))

unique_mails
```
