```python
some_list = [6, 6, 2, 8, 4, 0, 9, 4, 6, 8, 1]


def bubble_sort(arr):
    n = len(arr)  # 1

    for i in range(n - 1):  # n - 1
        swapped = False
        for j in range(0, n - 1 - i):  # n² <- Dominante Operation
            curr_val = arr[j]  # 1
            next_val = arr[j + 1]  # 1
            if curr_val > next_val:  # 1
                # swap
                # arr[j], arr[j + 1] = next_val, curr_val
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # 2
                swapped = True
        if not swapped:
            break
    return arr


print(some_list)
bubble_sort(some_list)
print(some_list)
```
