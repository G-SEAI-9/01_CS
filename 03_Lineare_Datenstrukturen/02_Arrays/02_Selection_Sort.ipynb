```python
some_numbers = [7, 3, 9, 5, 7, 1, 5, 8, -2, 7, -123, 7]


def selection_sort(arr):
    n = len(arr)  # O(1)

    for i in range(n - 1):  # O(n)
        min_index = i

        for j in range(i + 1, n):  # O(n²)
            if arr[j] < arr[min_index]:
                min_index = j
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]  # O(2)

    return arr
```

```python
print(some_numbers)
selection_sort(some_numbers)
print(some_numbers)
```
