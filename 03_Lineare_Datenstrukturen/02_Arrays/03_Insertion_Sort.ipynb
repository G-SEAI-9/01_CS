```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):  # O(n)
        insert_index = i
        current_value = arr[i]

        for j in range(i - 1, -1, -1):  # O(n²)
            if arr[j] > current_value:
                arr[j + 1] = arr[j]
                insert_index = j
            else:
                break

        arr[insert_index] = current_value

    return arr
```

```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        current_value = arr[i]

        j = i - 1

        while j >= 0 and arr[j] > current_value:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = current_value

    return arr
```

```python
numbers = [87, 34, 6, 1, -9, 42, 6, 9, 87, 35, -34]

print(numbers)
insertion_sort(numbers)
print(numbers)
```
