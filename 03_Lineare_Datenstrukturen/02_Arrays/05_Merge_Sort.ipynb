```python
my_list = [7, 3, 6, 5, 9]


def merge(arr, left, mid, right):
    n1 = mid - left + 1
    n2 = right - mid
    # L = [0] * n1
    # R = [0] * n2
    # for i in range(n1):
    #     L[i] = arr[left + 1]

    # for i in range(n2):
    #     R[i] = arr[mid + 1 + 1]

    L = arr[left : mid + 1]
    R = arr[mid + 1 : right + 1]  # O(n)

    i = 0
    j = 0
    k = left

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1


def merge_sort(arr, left, right, logging=False):  # O(n log n)
    if left < right:
        # Array spalten
        # mid = len(arr) // 2
        mid = (right + left) // 2
        if logging:
            print(
                f"Split: {arr[left:right+1]} into {arr[left:mid+1]} and {arr[mid+1:right+1]}"
            )
        merge_sort(arr, left, mid, logging)  # O(log n)
        merge_sort(arr, mid + 1, right, logging)

        if logging:
            print(f"Merging: {arr[left:mid+1]} + {arr[mid+1:right+1]}")
        merge(arr, left, mid, right)  # O(n)
        if logging:
            print(f"Result:  {arr[left:right+1]}")

    return arr


merge_sort(my_list, 0, len(my_list) - 1, logging=True)
```
