```python
my_str = "Hello"


# reversed() gibt die Zeichen eines Strings in umgekehrter Reihenfolge zurück.
# join() packt sie wieder zu einem String zusammen.
def rev_string(bla):
    return "".join(reversed(bla))


print(rev_string(my_str))
```

```python
# range(len(s) - 1, -1, -1) erzeugt Indizes vom letzten bis zum ersten Zeichen.
# In jeder Iteration wird das Zeichen an Position i ans Ergebnis angehängt.
def reverse_strings(s: str) -> str:
    result = ""
    for i in range(len(s) - 1, -1, -1):
        result += s[i]

    return result


print(rev_string("world"))
```

```python
# Jeder neue Buchstabe wird VOR das bisherige Ergebnis gesetzt.
# So entsteht der umgekehrte String Schritt für Schritt.
# Beispiel für "word": "" → "w" → "ow" → "row" → "drow"
def revert_string(text):
    result = ""
    for letter in text:
        result = letter + result

    return result


print(revert_string("world"))
print(revert_string("word"))
```

```python
# Solange der String noch Zeichen hat, wird das letzte Zeichen ans Ergebnis gehängt
# und dann aus dem String entfernt (s[:-1] = alles außer dem letzten Zeichen).
def reverse_while(s):
    result = ""
    while len(s) > 0:
        result += s[-1]
        s = s[:-1]

    return result


print(
    reverse_while(
        "uoy truh dna eil a llet annog reveNeybdoog yas annog reveNyrc uoy ekam annog reveNuoy tresed dna dnuora nur annog reveNnwod uoy tel annog reveNpu uoy evig annog reveN"
    )
)
```

```python
# REKURSION: Eine Funktion ruft sich selbst auf.
# Hier gibt es keine Abbruchbedingung — die Funktion läuft ewig
# und führt zu einem RecursionError (Stack Overflow).
def do_sth_recursively():
    print("Doing something")
    do_sth_recursively()  # ruft sich selbst auf → endlose Schleife!


do_sth_recursively()
```

```python
# global erlaubt, die Variable 'count' aus dem äußeren Scope zu verändern.
count = 0


def do_sth_recursively():
    global count
    count += 1

    # Abbruchbedingung (Base Case): Ohne sie läuft die Rekursion ewig.
    # Hier stoppen wir, sobald count den Wert 10 erreicht.
    if count == 10:
        print("Stopping")
        return

    print("Doing something")
    do_sth_recursively()  # rekursiver Aufruf — nächste Runde


do_sth_recursively()
```

```python
def reverse_recursively(s):

    # Base Case: Ein String mit einem (oder keinem) Zeichen ist bereits "umgekehrt".
    # Hier endet die Rekursion — ohne Base Case würde die Funktion "ewig" laufen.
    if len(s) <= 1:
        return s

    # Recursive Case: Nimm das letzte Zeichen und hänge den umgekehrten Rest dahinter.
    # Die Funktion löst ein kleineres Problem (s[:-1]) und baut daraus die Lösung.
    return s[-1] + reverse_recursively(s[:-1])

    # Gedankenmodell für "world":
    # "d" + reverse_recursively("worl")

    #     "l" + reverse_recursively("wor")

    #         "r" + reverse_recursively("wo")

    #           "o" + reverse_recursively("w")
    #                 → "w"  (Base Case!)


print(reverse_recursively("world"))
```

```python
# Dieses Beispiel zeigt den "Call Stack" — den Stapel offener Funktionsaufrufe.
# Jede Box wird geöffnet (hinzugefügt), bevor die nächste geöffnet wird.
# Erst wenn Box #1 erreicht ist (Base Case), werden die Boxen in umgekehrter
# Reihenfolge geschlossen — von innen nach außen.
def open_boxes(box_number):

    print(f"Opening Box #{box_number}")

    # Base Case: Box #1 hat keine weitere Box darin.
    if box_number == 1:
        print("Reached Box #1")
        return

    open_boxes(box_number - 1)  # öffne die nächst kleinere Box

    # Diese Zeile wird erst ausgeführt, NACHDEM der rekursive Aufruf zurückgekehrt ist.
    print(f"Closing Box #{box_number}")


open_boxes(3)
```

```python
# Baum-Rekursion: Eine Funktion ruft sich MEHRFACH selbst auf.
# Jede Person hat eine Mutter UND einen Vater — daher verzweigt der Baum.
# 'generation' zählt, wie tief wir im Baum sind.
def druck_stammbaum(name, generation=0):

    # Einrückung wächst mit jeder Generation — zeigt die Tiefe im Baum.
    indent = "  " * generation

    print(f"{indent} -> {name}")

    # Base Case: Nach 3 Generationen hören wir auf (sonst würde der Baum ewig wachsen).
    if generation == 3:
        return

    # Zwei rekursive Aufrufe: einmal für die Mutter, einmal für den Vater.
    # Dadurch verdoppelt sich die Anzahl der Aufrufe mit jeder Generation.
    druck_stammbaum(f"Mutter von {name}", generation + 1)
    druck_stammbaum(f"Vater von {name}", generation + 1)

    print(f"{indent} Zweig abgeschlossen {name}")


druck_stammbaum("Phil")
```
