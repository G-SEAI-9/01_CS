```python
nums = [5, 7, 3, 7, 2, 8, 0, 5, -8, 4, -2, 8, 3]


def partition(arr, left, right):
    pivot_value = arr[right]

    i = left - 1
    for j in range(left, right):
        if arr[j] <= pivot_value:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[right] = arr[right], arr[i + 1]

    return i + 1


def quicksort(arr, left, right):  # O(n log n)
    if left < right:
        pivot_index = partition(arr, left, right)

        quicksort(arr, left, right=(pivot_index - 1))
        quicksort(arr, left=(pivot_index + 1), right=right)

    return arr


quicksort(nums, 0, len(nums) - 1)  # [5, ... , 3 ] # [-8, ..., 2]
print(nums)
```

```python
import random

nums = [5, 7, 3, 7, 2, 8, 0, 5, -8, 4, -2, 8, 3]


def partition(arr, left, right):
    pivot_index = random.randint(left, right)
    arr[pivot_index], arr[right] = arr[right], arr[pivot_index]

    pivot_value = arr[right]

    i = left - 1
    for j in range(left, right):
        if arr[j] <= pivot_value:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[right] = arr[right], arr[i + 1]

    return i + 1


def quicksort(arr, left, right):  # O(n log n)
    if left < right:
        pivot_index = partition(arr, left, right)

        quicksort(arr, left, right=(pivot_index - 1))
        quicksort(arr, left=(pivot_index + 1), right=right)

    return arr


quicksort(nums, 0, len(nums) - 1)  # [5, ... , 3 ] # [-8, ..., 2]
print(nums)
```
