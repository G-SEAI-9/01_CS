```python
def binary_search(arr, target):  # O(log n)
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if target == arr[mid]:
            print(f"Gefunden {target} an Index {mid}")
            return mid
        elif target > arr[mid]:
            print(f"Target größer als {arr[mid]} an {mid}")
            left = mid + 1
        else:
            print(f"Target kleiner als {arr[mid]} an {mid}")
            right = mid - 1

    return -1
```

```python
binary_search(nums, 42)
```
