```python
nums = [5, 8, 2, 5, 9, 4, 7]

target = 5


print(nums.index(target))
```

```python
def linear_search(arr, target):  # O(n)
    for i in range(len(arr)):
        if arr[i] == target:
            return i

    return -1


linear_search(nums, 5)
```

```python
def linear_search(arr, target):
    for index, value in enumerate(arr):
        if value == target:
            return index

    return -1


linear_search(nums, 4)
```
