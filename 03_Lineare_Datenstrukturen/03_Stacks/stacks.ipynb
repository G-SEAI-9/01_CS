```python
class Stack:
    def __init__(self):
        self.items = []
        self.length = 0

    def push(self, item):
        self.items.append(item)
        self.length += 1

    def pop(self):
        if self.is_empty():
            return None

        self.length -= 1
        return self.items.pop()

    def size(self):
        return self.length

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        return self.items[-1]
```

```python
my_stack = Stack()
# print(my_stack)

my_stack.push("Löffel")
my_stack.push("Fahrradkettenöl")

print(my_stack.size())
print(my_stack.peek())

last = my_stack.pop()
print(last)
print(my_stack.size())
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.pop()
last = my_stack.pop()
print(last)
```

```python
def fakultät(n):
    if n <= 1:
        return 1

    return n * fakultät(n - 1)


print(fakultät(5))
# print(fakultät(6))
```

```python
def fakultät_iterativ(n):
    stack = Stack()

    while n > 1:
        stack.push(n)
        n -= 1

    result = 1

    while not stack.is_empty():
        result *= stack.pop()

    return result
```

```python
print(fakultät_iterativ(5))
print(fakultät_iterativ(6))
```
