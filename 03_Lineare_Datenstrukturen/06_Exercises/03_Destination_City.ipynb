```python
"""
You are given an array, paths, where each element paths[i] = [cityAᵢ, cityBᵢ] represents a direct route from cityAᵢ to cityBᵢ.
Your task is to identify the final city on the route — the one that has no outgoing path to any other city.
The structure of the routes is guaranteed to be a simple chain without loops, so there will be exactly one such destination.
https://leetcode.com/problems/destination-city/description/
"""
```

```python
paths = [
    ["London", "New York"],
    ["Lima", "Sao Paulo"],
    ["New York", "Lima"],
]
# Output: "Sao Paulo"
#
# paths = [["B", "C"], ["D", "B"], ["C", "A"]]
# Output: "A"

# paths = [["A", "Z"]]
# Output: "Z"


def destCity(paths):  # O(n)
    # Set Comprehension: Alle Startstädte in einem Schritt als Set einsammeln.
    # Ein Set speichert jeden Wert nur einmal und prüft Mitgliedschaft in O(1).
    starts = {path[0] for path in paths}  # O(n)
    print(starts)

    # Alle Zielstädte ebenfalls als Set
    ends = {path[1] for path in paths}  # O(n)

    # Die Stadt, die als Ziel vorkommt, aber nie als Start → das ist die Endstation
    for end in ends:  # O(n)
        if end not in starts:  # O(1)!
            return end


print(destCity(paths))
```

```python
def find_final_city(paths):  # Sammelboxen erstellen - O(2n) -> O(n)
    starts = set()
    destinations = set()

    # Jeden Pfad aufteilen: linke Stadt = Start, rechte Stadt = Ziel.
    # Durch Unpacking (a, b) greifen wir direkt auf beide Elemente zu.
    for a, b in paths:  # städte aufteilen - O(n)
        starts.add(a)  # Start stadt
        destinations.add(b)  # Ziel Stadt

    # Die Stadt, die nur Ziel und nie Start ist, ist die Endstadt
    # `not in starts` ist O(1), weil starts ein Set ist
    for city in destinations:  # O(n)
        if city not in starts:  # O(1)
            return city
```

```python
def find_final_city(paths):
    starts = set()
    destinations = set()

    for a, b in paths:  # städte aufteilen - O(n)
        starts.add(a)  # Start stadt
        destinations.add(b)  # Ziel Stadt

    # Mengen-Differenz: destinations - starts liefert alle Elemente,
    # die in destinations vorkommen, aber nicht in starts.
    # Da es garantiert genau eine Endstadt gibt, enthält das Ergebnis-Set
    # nur ein Element — .pop() holt es heraus.
    return (destinations - starts).pop()


print(find_final_city(paths))
```

```python
def destCity(paths):
    # Alle Startstädte in einer einfachen Liste sammeln
    start_paths_list = []
    for contition in paths:
        start_paths_list.append(contition[0])

    # Für jede Zielstadt prüfen, ob sie auch als Startstadt vorkommt.
    # Achtung: `in` auf einer Liste durchsucht ca. alle Einträge → O(n) pro Prüfung.
    # Die Gesamtlaufzeit wird dadurch O(n²) — mit einem Set wäre es O(n).
    for condition in paths:  # O(n)
        if condition[1] not in start_paths_list:  # O(n) <- bei Listen
            return condition[1]
```

```python
paths = [
    ["London", "New York"],
    ["Lima", "Sao Paulo"],
    ["New York", "Lima"],
]


def destCity(paths):  # O(n) (Leicht ineffiziender als Setdifferenz)
    # paths ist eine Liste von [Start, Ziel]-Paaren.
    # dict(paths) wandelt sie direkt in ein Dictionary um:
    # {"London": "New York", "New York": "Lima", "Lima": "Sao Paulo"}
    # Jede Startstadt zeigt damit auf ihre nächste Zielstadt.
    destinations_map = dict(paths)  # O(n)

    # Wir starten an der Zielstadt des ersten Eintrags und reisen von dort weiter.
    curr_place = paths[0][1]

    # Solange die aktuelle Stadt selbst noch Startpunkt eines Pfades ist, weiterreisen.
    # Die Schleife endet, wenn curr_place nicht mehr als Schlüssel im Dictionary steht
    # → das ist die Endstation ohne ausgehenden Pfad.
    while curr_place in destinations_map:  # O(n)
        curr_place = destinations_map.get(curr_place)

    return curr_place


print(destCity(paths))
```

```python
cases = [
    (
        [[["London", "New York"], ["New York", "Lima"], ["Lima", "Sao Paulo"]]],
        "Sao Paulo",
    ),
    ([[["B", "C"], ["D", "B"], ["C", "A"]]], "A"),
    ([[["A", "Z"]]], "Z"),
]


def test(fn, cases):
    for input, expected in cases:
        output = fn(*input)
        try:
            assert output == expected
            print(f"Test succeeded: {input} -> {output}")
        except AssertionError:
            print(f"Test failed: expected {expected}, got {output}")


test(destCity, cases)
```
