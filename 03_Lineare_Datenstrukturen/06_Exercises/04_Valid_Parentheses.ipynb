```python
"""
You have a string s composed only of the characters (, ), {, }, [ and ]. Determine if the bracket arrangement in this string is valid. Specifically, a valid string must meet the following criteria:

Each opening bracket has a corresponding closing bracket of the same type.
Brackets are closed in the correct order.
Every closing bracket matches an earlier opening bracket of the same type.
https://leetcode.com/problems/valid-parentheses/description/
"""
```

```python
s = "([])"
# Out: True
#
# s = "([)]"
# Out: False

# s = "]"
# Out: False

s = "((((((((((((((("  # Worst Case: nur öffnende Klammern → Stack füllt sich komplett

# Dictionary: jede schließende Klammer zeigt auf ihre passende öffnende
pairs = {")": "(", "]": "[", "}": "{"}


def isValid(s: str) -> bool:  # O(n)
    # Der Stack speichert alle öffnenden Klammern, die noch nicht geschlossen wurden
    stack = []  # mem: O(n) - Stack könnte so groß werden wie der Input s

    for char in s:  # O(n) — wir gehen jeden Charakter genau einmal durch
        if char in "([{":  # öffnende Klammer → auf den Stack legen
            stack.append(char)
        else:
            # schließende Klammer gefunden — jetzt prüfen ob sie passt

            if len(stack) == 0:
                # Stack ist leer: es gibt keine offene Klammer zum Schließen
                return False

            if stack[-1] != pairs[char]:
                # Die zuletzt geöffnete Klammer passt nicht zur aktuellen schließenden
                # Beispiel: stack hat '(' oben, aber wir sehen ']' → ungültig
                return False

            # Passendes Paar gefunden → öffnende Klammer vom Stack entfernen
            stack.pop()  # O(1)

    # Am Ende muss der Stack leer sein — sonst gibt es ungeschlossene Klammern
    return len(stack) == 0


print(isValid(s))
```

```python
# Alternative Implementierung: gleiche Logik, aber umgekehrte Denkrichtung
# Statt "ist der zuletzt geöffnete der passende?" prüfen wir:
# "ist die schließende Klammer bekannt, und passt der Stack-Top dazu?"
parens_map = {")": "(", "]": "[", "}": "{"}


def isValid(s: str) -> bool:
    stack = []
    for char in s:
        if char in parens_map:
            # char ist eine schließende Klammer
            # Stack-Top muss die passende öffnende Klammer sein
            if len(stack) > 0 and stack[-1] == parens_map.get(char):
                stack.pop()  # Paar komplett → beide entfernen
            else:
                # Stack leer oder falsche öffnende Klammer oben
                return False
        else:
            # char ist eine öffnende Klammer → für später merken
            stack.append(char)

    # Alle Klammern wurden korrekt gepaart, wenn der Stack leer ist
    return len(stack) == 0
```

```python
# Testfälle: jeder Eintrag ist ([input], erwartetes_ergebnis)
cases = [
    (
        ["([])"],
        True,  # verschachtelt, aber korrekt geschlossen
    ),
    (["([)]"], False),  # falsche Reihenfolge beim Schließen
    (["]"], False),  # schließende Klammer ohne öffnende
]


def test(fn, cases):
    for inp, expected in cases:
        output = fn(*inp)
        try:
            assert output == expected
            print(f"Test succeeded: {inp} -> {output}")
        except AssertionError:
            print(f"Test failed: expected {expected}, got {output}")


test(isValid, cases)
```
