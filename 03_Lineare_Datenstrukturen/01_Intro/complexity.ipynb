```python
# Wenn mein Input anwächst, wie wächst mein Rechenaufwand? -> Zeitkomplexität
# Wenn mein Input anwächst, wie wächst mein Speicherbedarf? -> space complexity


def find_largest(arr):  # O(n²) -> Quadratisch
    result = []
    n = len(arr)  # 100_000

    for i in range(n):  # n

        is_largest = True  # 1

        for j in range(n):  # n²
            result.append(j)
            if arr[j] > arr[i]:  # 1

                is_largest = False  # 1

                break  #

        if is_largest:  # 1
            return arr[i]


# 3 + 2n + n²

# n = 1 -> 3 + 2 + 1
# n = 100  -> 3 + 200 + 10_000


numbers = [3, 1, 4, 1, 5, 9, 1, 1, 1, 1, 1, 1]
largest = find_largest(numbers)
print("Largest number: ", largest)
```

```python
def find_largest(arr):  # O(n) -> Linear

    max_num = arr[0]  # 1

    for i in range(1, len(arr)):  # n - 1
        if arr[i] > max_num:  # 1
            max_num = arr[i]  # 1

    return max_num


# 9 -> 8; 9_000_000 -> 8_999_999

numbers = [3, 1, 4, 1, 5, 9]
largest = find_largest(numbers)
print("Largest number: ", largest)
```
