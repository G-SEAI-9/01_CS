```python
# Eine Node ist das Grundbaustein einer Linked List.
# Jede Node speichert einen Wert und einen Zeiger auf die nächste Node.
class Node:
    def __init__(self, value, next=None):
        self.value = value  # die eigentlichen Daten, die diese Node enthält
        self.next = next  # Zeiger auf die nächste Node; None bedeutet: kein Nachfolger
```

```python
class LinkedList:
    def __init__(self):
        # head zeigt auf die erste Node der Liste; None = leere Liste
        self.head = None
        self.length = 0

    def append(self, value):  # O(n)
        new_node = Node(value)

        if not self.head:
            # Sonderfall: leere Liste – die neue Node wird direkt zum head
            self.head = new_node
        else:
            # Wir müssen die letzte Node finden, um eine neue anzuhängen.
            # Dafür laufen wir die gesamte Liste durch – daher O(n).
            current_node = self.head
            while current_node.next:
                current_node = current_node.next

            # current_node ist jetzt die letzte Node; wir hängen die neue an
            current_node.next = new_node

        self.length += 1

    def pop(self):  # O(n)
        if not self.head:
            return None

        self.length -= 1

        if not self.head.next:
            # Sonderfall: nur ein Element in der Liste
            val = self.head.value
            self.head = None
            return val

        # Wir brauchen die vorletzte Node, um deren next-Zeiger auf None zu setzen.
        # Dafür behalten wie die prev_node im Hinterkopf.
        current_node = self.head
        prev_node = None

        while current_node.next:
            prev_node = current_node
            current_node = current_node.next

        if prev_node:
            val = current_node.value
            # Die vorletzte Node von der letzten entkoppeln – die letzte wird damit unerreichbar (und vom Garbage Collector gelöscht)
            prev_node.next = None
            return val

    def __str__(self):
        # Nur zur Visualisierung
        if not self.head:
            return "Empty List"

        nodes = []
        curr = self.head
        while curr:
            nodes.append(f"[{curr.value}]")
            curr = curr.next
        return " -> ".join(nodes) + " -> None"
```

```python
ll = LinkedList()

print(ll)

ll.append(42)
print(ll)

ll.append("Hello")
print(ll)

ll.append("World")
print(ll)

last = ll.pop()
print(last)
print(ll)
```

```python
# Eine Queue auf Basis einer Linked List mit einem zusätzlichen tail-Zeiger.
# Das ermöglicht enqueue und dequeue in O(1) – ohne die Liste durchlaufen zu müssen.
class QueueLL:
    def __init__(self):
        self.head = (
            None  # Zeiger auf das vorderste Element (wird als nächstes rausgenommen)
        )
        self.tail = (
            None  # Zeiger auf das hinterste Element (neues Element wird hier angehängt)
        )
        self.length = 0

    def enqueue(self, value):  # O(1)
        new_node = Node(value)

        if not self.tail:
            # Leere Queue: head und tail zeigen beide auf den einzigen Node
            self.head = new_node
            self.tail = new_node
        else:
            # Wir kennen das letzte Element direkt über self.tail – kein Durchlaufen nötig.
            # Die bisherige tail-Node zeigt jetzt auf die neue, und tail rückt vor.
            self.tail.next = new_node
            self.tail = new_node

        self.length += 1

    def dequeue(self):  # O(1)
        if not self.head:
            return None

        dequeued_val = self.head.value
        # head rückt zur nächsten Node vor – das bisherige head ist damit raus
        self.head = self.head.next

        if not self.head:
            # Die Queue ist jetzt leer; tail muss ebenfalls zurückgesetzt werden,
            # damit kein veralteter Zeiger auf die entfernte Node übrig bleibt.
            self.tail = None

        self.length -= 1
        return dequeued_val

    def __str__(self):
        # Nur zur Visualisierung
        nodes = []
        curr = self.head
        while curr:
            nodes.append(f"[{curr.value}]")
            curr = curr.next
        return "(Head) " + " <- ".join(nodes) + " (Tail)"
```

```python
q = QueueLL()

print(q)

q.enqueue("Druckauftrag 1")
q.enqueue("Druckauftrag 2")
print(q)

q.dequeue()
print(q)
```
