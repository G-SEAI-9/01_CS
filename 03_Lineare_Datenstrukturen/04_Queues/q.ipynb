```python
class Queue:
    def __init__(self) -> None:
        self.items = []

    def enqueue(self, item):  # push
        self.items.append(item)

    def dequeue(self):  # .pop(0)
        if self.is_empty():
            return None

        return self.items.pop(0)  # O(n)

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def peek(self):
        if self.is_empty():
            return None
        return self.items[0]
```

```python
warteschlange = Queue()

warteschlange.enqueue("Tomas")
warteschlange.enqueue("Eric")

print(warteschlange.size())

kunde = warteschlange.dequeue()
print(kunde)
print(warteschlange.size())

warteschlange.enqueue("Katharina")
warteschlange.enqueue("Nicole")
warteschlange.enqueue("Michael")

while not warteschlange.is_empty():
    kunde = warteschlange.dequeue()
    print(kunde)
```

```python
class Queue:
    def __init__(self) -> None:
        self.items = []
        self.head = 0

    def enqueue(self, item):  # push
        self.items.append(item)

    def dequeue(self):
        if self.is_empty():
            return None

        dequeued_item = self.items[self.head]
        self.head += 1
        return dequeued_item

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def peek(self):
        if self.is_empty():
            return None
        return self.items[0]
```
