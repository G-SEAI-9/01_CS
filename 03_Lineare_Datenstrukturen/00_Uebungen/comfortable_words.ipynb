```python
lefty = {"q", "w", "e", "r", "t", "a", "s", "d", "f", "g", "z", "x", "c", "v", "b"}
righty = {"y", "u", "i", "o", "p", "h", "j", "k", "l", "n", "m"}


def cozy_writing(word):
    last_hand = None

    for letter in word:  # O(n)
        if letter in lefty:  # O(1)
            if last_hand == "left":
                return False
            else:
                last_hand = "left"
        if letter in righty:  # O(1)
            if last_hand == "right":
                return False
            else:
                last_hand = "right"

    return True


print(cozy_writing("yams"))
print(cozy_writing("test"))
```

```python
def is_comfortable(word):
    left_hand = set("qwertasdfgzxcvb")
    for i in range(len(word) - 1):
        current_is_left = word[i] in left_hand
        next_is_left = word[i + 1] in left_hand
        if current_is_left == next_is_left:
            return False

    return True


print(is_comfortable("yams"))
print(is_comfortable("test"))
```

```python
left = {"q", "w", "e", "r", "t", "a", "s", "d", "f", "g", "z", "x", "c", "v", "b"}
right = {"y", "u", "i", "o", "p", "h", "j", "k", "l", "n", "m"}


def comfortable_word(word):
    is_left = word[0] in left

    for letter in word:
        lookup = left if is_left else right
        if letter not in lookup:
            return False
        is_left = not is_left

    return True
```

```python
left = {"q", "w", "e", "r", "t", "a", "s", "d", "f", "g", "z", "x", "c", "v", "b"}
right = {"y", "u", "i", "o", "p", "h", "j", "k", "l", "n", "m"}


def comfortable_word(word):
    is_left = word[0] in left
    evens = set(word[0::2])  # O(n)
    odds = set(word[1::2])  # O(n)
    return (
        evens.issubset(left) and odds.issubset(right)  # O(1)
        if is_left
        else evens.issubset(right) and odds.issubset(left)
    )


print(comfortable_word("yams"))
print(comfortable_word("test"))
```
