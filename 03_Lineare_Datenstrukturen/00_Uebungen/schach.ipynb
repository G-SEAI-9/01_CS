```python
# Aufgabe 2:
# - printe das Schachbrett so formatiert, wie im angehГӨngten Screenshot
# - fГјge der Originalvariablen keine zusГӨtzlichen Zeichen hinzu, sondern lГ¶se die Aufgabe Гјber Stringformatierung und loops

# for row in chess_board:
#   print(row)


def print_board():
    print("  a b c d e f g h")
    print("  _______________")

    for index, row in enumerate(chess_board):
        row_number = 8 - index
        print(f"{row_number}|{" ".join(row)}|{row_number}")

    print("  _______________")
    print("  a b c d e f g h")
```

```python
# Aufgabe 4:
# - erstelle eine Funktion, um einen Spielzug durchzufГјhren
# - erstelle eine Funktion, um den aktuellen Zustand des Schachbrett auszugeben


def move_piece(start, target):
    """Verschiebt eine Figur von einem Startfeld auf ein Zielfeld."""
    start_column = ord(start[0].lower()) - ord("a")  # e2
    start_row = 8 - int(start[1])

    target_column = ord(target[0].lower()) - ord("a")
    target_row = 8 - int(target[1])

    chess_board[target_row][target_column] = chess_board[start_row][start_column]
    chess_board[start_row][start_column] = " "


def move(board, from_row, from_col, to_row, to_col):
    board[to_row][to_col] = board[from_row][from_col]
    board[from_row][from_col] = " "
```

```python
# Bonus Bonus:
# - erstelle einen (einfachen) Spielloop, der abwechselnd WeiГҹ und Schwarz ziehen lГӨsst und das Spielfeld nach jedem Zug neu ausgibt
# - (lass die Zugvalidation, also die PrГјfung, ob eine Figur einen bestimmten Zug machen darf, hier aus)

chess_board = [
    ["вҷң", "вҷһ", "вҷқ", "вҷӣ", "вҷҡ", "вҷқ", "вҷһ", "вҷң"],
    ["вҷҹ", "вҷҹ", "вҷҹ", "вҷҹ", "вҷҹ", "вҷҹ", "вҷҹ", "вҷҹ"],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    ["вҷҷ", "вҷҷ", "вҷҷ", "вҷҷ", "вҷҷ", "вҷҷ", "вҷҷ", "вҷҷ"],
    ["вҷ–", "вҷҳ", "вҷ—", "вҷ•", "вҷ”", "вҷ—", "вҷҳ", "вҷ–"],
]
```

```python
def game_loop():
    """LГӨsst WeiГҹ und Schwarz abwechselnd ziehen."""
    current_player = "WeiГҹ"

    while True:
        print_board()
        print(f"{current_player} ist am Zug.")

        start = input("Startfeld eingeben oder 'quit' zum Beenden: ")

        if start.lower() == "quit":
            break

        target = input("Zielfeld eingeben: ")
        move_piece(start, target)

        if current_player == "WeiГҹ":
            current_player = "Schwarz"
        else:
            current_player = "WeiГҹ"
```
