```python
#  >>> fizzbuzz(20)

# 1, 2, fizz, 4, buzz, fizz, 7, 8, fizz, buzz, 11, fizz, 13, 14, fizzbuzz, 16, 17, fizz, 19, buzz
# Ist die Zahl teilbar durch 3 - "fizz"
# Ist die Zahl teilbar durch 5 - "buzz"
# Ist die Zahl teilbar durch 3 und 5 - "fizzbuzz"
```

```python
item = int(input("Gib eine Zahl ein: "))

for num in range(1, item + 1):
    match num:
        case _ if num % 3 == 0 and num % 5 == 0:
            print("fizzbuzz")
        case _ if num % 3 == 0:
            print("fizz")
        case _ if num % 5 == 0:
            print("buzz")
        case _:
            print(num)
```

```python
for i in range(1, 101):
    if i % 15 == 0:
        print("fizzbuzz")
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
    else:
        print(i)
```

```python
for i in range(1, 21):
    match (i % 3, i % 5):
        case (0, 0):
            print("fizzbuzz")
        case (0, _):
            print("fizz")
        case (_, 0):
            print("buzz")
        case _:
            print(i)
```
